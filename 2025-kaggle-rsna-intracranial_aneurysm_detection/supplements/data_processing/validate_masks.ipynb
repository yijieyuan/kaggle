```python
import numpy as np
from pathlib import Path

mask_dir = Path(r"E:\kaggle-rsna-data_processing3\mask_256")

for npy_file in mask_dir.glob("*.npy"):
    mask = np.load(npy_file)
    unique_labels = np.unique(mask)
    
    # Check if all labels 1-13 are present
    expected_labels = set(range(1, 14))
    actual_labels = set(unique_labels)
    
    if not expected_labels.issubset(actual_labels):
        missing = expected_labels - actual_labels
        print(f"{npy_file.name}: Missing labels {missing}")

print("Check complete!")
```
